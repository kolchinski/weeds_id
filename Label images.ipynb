```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
```

```python
def crosshatch_image(image_name):
    img = Image.open('deep_weeds_dataset/images/{}.jpg'.format(image_name))
    margin = (256-224)//2

    im_array = np.asarray(img)[margin:-margin, margin:-margin, :].copy()
    print(im_array.shape)

    for i in range(6):
        spacing = 224//7 * (i + 1)
        im_array[spacing, :, 0:2] = 255.
        im_array[:, spacing, 0:2] = 255.


    plt.imshow(im_array)
    plt.show()

crosshatch_image('20170718-133830-2')
```
